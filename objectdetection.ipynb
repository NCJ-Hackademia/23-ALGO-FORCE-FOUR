```python
from ultralytics import YOLO
import pyttsx3
import cv2
import os
import time

# Initialize the text-to-speech engine
try:
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)  # Adjust speech rate for clarity
except Exception as e:
    print(f"Error initializing pyttsx3: {e}")
    exit()

# Load the pretrained YOLOv8n model
model_path = r'C:\Users\User\amrik\yolov8n.pt'  # Your specified path
if not os.path.exists(model_path):
    print(f"Error: Model file {model_path} does not exist.")
    print("Please ensure yolov8n.pt is in the specified directory or download it.")
    exit()

try:
    model = YOLO(model_path)
except Exception as e:
    print(f"Error loading YOLO model: {e}")
    exit()

# Open the built-in laptop camera (index 0)
cap = cv2.VideoCapture(0)

# Check if the camera is opened correctly
if not cap.isOpened():
    print("Error: Could not open the built-in camera. Ensure it is enabled and not in use by another application.")
    exit()

# Optional: Set frame resolution (adjust if needed)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

print("Built-in camera opened. Point the camera at objects to detect them.")
print("Press 'q' to quit the camera feed.")

# To limit speech frequency (e.g., every 3 seconds)
last_speech_time = 0
speech_interval = 3  # Seconds between speech outputs

while True:
    # Read a frame from the camera
    ret, frame = cap.read()
    
    if not ret:
        print("Error: Failed to capture frame from the camera.")
        break
    
    # Run object detection on the frame
    try:
        results = model(frame)
    except Exception as e:
        print(f"Error running YOLO model on frame: {e}")
        break
    
    # To store unique detected class names in this frame
    detected_objects = set()
    
    # Process the results
    for r in results:
        for box in r.boxes:
            # Get the class index and name
            cls = int(box.cls[0])
            class_name = model.names[cls]
            detected_objects.add(class_name)
    
    # Speak unique detected objects if enough time has passed
    current_time = time.time()
    if detected_objects and (current_time - last_speech_time) >= speech_interval:
        for obj in detected_objects:
            print(f"Detected: {obj}")
            engine.say(f"I see a {obj}")
        try:
            engine.runAndWait()  # Speak the queued phrases
            last_speech_time = current_time
        except Exception as e:
            print(f"Error in text-to-speech: {e}")
    
    # If no objects detected
    if not detected_objects:
        print("No objects detected in this frame.")
    
    # Visualize the results on the frame (draws bounding boxes and labels)
    annotated_frame = results[0].plot()
    
    # Display the annotated frame
    cv2.imshow("YOLOv8 Built-in Camera Detection", annotated_frame)
    
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()
```
